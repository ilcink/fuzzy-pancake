# How to create an R dataframe (df)?

```r
# Create a, b, c, d variables using the operation "c": Combine Values into a Vector or List ( https://www.rdocumentation.org/packages/base/versions/3.6.2/topics/c )
a <- c("AAAGATGAGCAGGTCA", 
"AAGTCTGGTGTTTGTG", 
"ACACCAAGTTCATGGT",
"ACCAGTAGTTTGCATG",
"CAGCTGGCACAGGTTT",
"CACCACTGTCAACATC", 
"CGGTTAACAGGCTGAA",
"CGGTTAATCTTACCTA",  
"CTAATGGAGGTTCCTA",
"CTTGGCTAGAGGTAGA")
b <- c("HC","HC","HC",
"AA","AA","AA","AA",
"WA","WA","WA")
c <- c("T1","T2","T3",
"T3","T2","T1","T2",
"T3","T2","T4")
d <- c("TRA:CALSGIYNQGGKLIF,TRB:CASRSEGQEYNEQFF",
"",
"TRA:CGAVYGPSGSRLTF,TRB:CAFSASLAKNIQYF", 
"TRA:CAVRDMRNTDKLIF,TRB:CASSQFPGAGTEAFF",
"TRA:CAVRDMRNTDKLIF,TRB:CASSQFPGAGTEAFF",
"TRA:CAVRDMRNTDKLIF,TRB:CASSQFPGAGTEAFF",
"TRA:CAPGGPSSNTGKLIF,TRB:CASSGQGDNQPQHF",
"",
"TRA:CIVIQGGQGNLIF,TRB:CSVASAGWIGENIQYF",
"")
e <- c("",
"",
"",
"",
"",
"",
"",
"TRD:CALGVFGGLGDIRARPLIF,TRG:CATWGYSSDWIKTF",
"",
"TRD:CALGVFGGLGDIRARPLIF,TRG:CATWGYSSDWIKTF")
# Join the variables to create a data frame
df <- data.frame(a,b,c,d,e)
df
```

```r
#name the columns
names(df) <- c('cell_barcode', 'condition', 'cluster', 'TCR_ab', "TCR_gd")
df
```

```r
#renaming columns could be done as follows:
names(df)[names(df) == 'cluster'] <- 'cell_type'
df
```

```r
#add a new column with merged TCR_ab and TCR_gd values
library(dplyr)
df<-mutate(df, TCR_ab_gd = paste0(TCR_ab, "", TCR_gd))
df
```

# How to select the most frequent value in a df column?

```r
#add frequency: count how many times the same value (TCR_ab_gd clonotype) is encountered in the dataset
df2 <- transform(df, Frequency = ave(seq(nrow(df)), TCR_ab_gd, FUN = length))
df2
```

```r
#count the total amount of rows (cells / "cell_barcode") per condition
df3 <- transform(df2, Cell_count_from_one_condition = ave(seq(nrow(df2)), condition, FUN = length))
df3
```

```r
#clarify the proportion of each clonotype within one condition
df4 <- df3 %>% mutate(Percentage = Frequency*100 / Cell_count_from_one_condition)
df4
```

```r
#sort the cells in the descending order based on the clonotype prevalence
df5 <- df4 %>% arrange(desc(Frequency))
df5
```

```r
#Missing values are represented by the symbol NA (not available). Replace the missing value "" within a column "TCR_ab_gd" with "NA"
df5["TCR_ab_gd"][df5["TCR_ab_gd"] == ""] <- NA
df5
```

```r
#Select the clonotypes of interest, e.g. top clonotypes, those that are present more than once, and combine the rest rows as seperate items
df5$TCR_top <- ifelse(grepl(pattern = c("TRA:CAVRDMRNTDKLIF,TRB:CASSQFPGAGTEAFF"), df5$TCR_ab_gd), "AA_TRA_CAVRDMRNTDKLIF_TRB_CASSQFPGAGTEAFF",
                ifelse(grepl(pattern = c("TRD:CALGVFGGLGDIRARPLIF,TRG:CATWGYSSDWIKTF"), df5$TCR_ab_gd), "WA_TRD_CALGVFGGLGDIRARPLIF_TRG_CATWGYSSDWIKTF",
                    ifelse(!is.na(df5$TCR_ab_gd), "other_clonotype", "no_TCR")))
df5
```

```r
#clean up the table to avoid any confusion with regard to the clonotype frequencies and percentages for the rows with "other_clonotype" and "no_TCR"
df5["Frequency"][df5["TCR_top"] == "other_clonotype"] <- ""
df5["Frequency"][df5["TCR_top"] == "no_TCR"] <- ""
df5["Percentage"][df5["TCR_top"] == "other_clonotype"] <- ""
df5["Percentage"][df5["TCR_top"] == "no_TCR"] <- ""
df5
```

# Conditional subseting of the df either by column or by row

# A. Subsetting the df by column condition

# E.g. Subset the data so that only the rows containing the value "AA" in the column "condition" are preserved

```r
#Conduct a conditional subseting based on the value "AA" within the column "condition" across all df rows
#subset by column condition
df5_AA<-df5[df5$condition == 'AA',]
df5_AA
```

```r
#turn the values of a certain column into the rownames
rownames(df5_AA) <- df5_AA$"cell_barcode"
df5_AA
```

```r
#the column can be removed as follows
df5_AA$"cell_barcode"  <- NULL
df5_AA
```
